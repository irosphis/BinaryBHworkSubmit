```python
import h5py

# 打开 HDF5 文件
with h5py.File("hdf5/BinaryBHPlot_000000.3d.hdf5", "r") as f:
    # 打印文件结构
    def print_structure(name, obj):
        print(name)
    f.visititems(print_structure)

```

```python
import yt


file_path = "hdf5/BinaryBHPlot_000000.3d.hdf5"
ds = yt.load(file_path)

slc = yt.SlicePlot(ds, "z", "chi", center=[32, 32, 0])
slc.set_cmap("chi", "RdBu_r")
slc.set_width((64, 64)) 
```

```python
import yt
import glob
from PIL import Image
import os


files = sorted(glob.glob("hdf5/BinaryBHPlot_*.3d.hdf5"))


png_files = []
for i, file in enumerate(files[::5]): 
    ds = yt.load(file)
    slc = yt.SlicePlot(ds, "z", "chi", center=[32, 32, 0])
    slc.set_cmap("chi", "RdBu_r")
    slc.set_width((64, 64))
    output = f"fig/chi_t{i:03d}.png"
    slc.save(output)
    png_files.append(output)


images = [Image.open(png) for png in png_files]
images[0].save(
    "chi_animation.gif",
    save_all=True,
    append_images=images[1:],
    duration=200,  # 每帧 200ms
    loop=0
)


for png in png_files:
    os.remove(png)
```

```python
import yt
import glob
from PIL import Image
import os


files = sorted(glob.glob("hdf5/BinaryBHPlot_*.3d.hdf5"))[:300]


png_files = []
for i, file in enumerate(files[::1]): 
    ds = yt.load(file)
    slc = yt.SlicePlot(ds, "z", "chi", center=[32, 32, 0])
    slc.set_cmap("chi", "RdBu_r")
    slc.set_width((64, 64))
    output = f"fig/chi_t{i:03d}.png"
    slc.save(output)
    png_files.append(output)


images = [Image.open(png) for png in png_files]
images[0].save(
    "chi_animation.gif",
    save_all=True,
    append_images=images[1:],
    duration=200,
    loop=0
)

for png in png_files:
    os.remove(png)
```

```python
import yt
import glob
import matplotlib.pyplot as plt
from matplotlib.animation import FFMpegWriter

files = sorted(glob.glob("hdf5/BinaryBHPlot_*.3d.hdf5"))[:300]

fig, ax = plt.subplots(figsize=(8, 8))
ax.set_title("χ Field Evolution")
ax.set_axis_off()

writer = FFMpegWriter(fps=10, metadata={"title": "BH Merger"})

ds_sample = yt.load(files[0])
slc_sample = ds_sample.slice("z", 0)
frb_sample = slc_sample.to_frb((64, 'code_length'), 64)


with writer.saving(fig, "chi_animation.mp4", dpi=100):
    for i, file in enumerate(files[::1]):
        ds = yt.load(file)
        slc = ds.slice("z", 0)
        
        frb = slc.to_frb((64, 'code_length'), 64)
        arr = frb["chi"]
        
        img = ax.imshow(arr, extent=(-32, 32, -32, 32), cmap="RdBu_r")
        
        writer.grab_frame()
        ax.clear()
```
